# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.

```python
#Sorting Goal difference on descending order. This will make it easier to retrieve max and min goal difference.

def Goal_diff_count():
    return pd.DataFrame({'Team':laliga['Team'], 'Goal_diff_count': laliga['GoalsFor'].astype(int) - laliga['GoalsAgainst'].astype(int)}).sort_values(by=['Goal_diff_count'],ascending=False) 
    
```

```python
#function output with Team and Goal difference:
Goal_diff_count()
```

```python
Goal_diff_count().head(1)
```

### team that has min goal difference:

```python
Goal_diff_count().tail(1)
```

# Create a new column with the name “Winning Percent” and append it to the data set

```python
laliga['Winning Percent'] = (laliga['GamesWon'].astype(int) / laliga['GamesPlayed'].astype(int))*100
```

```python
laliga.head()
```

## Print the top 5 teams which have the highest Winning percentage

```python
laliga['Winning Percent'].fillna(0,inplace = True)
laliga[['Team', 'Winning Percent']].sort_values(by='Winning Percent', ascending=False).head(5)
```

## Group teams based on their “Best position” and print the sum of their points for all positions

```python
print(laliga.groupby('BestPosition')['Points'].sum().reset_index(name ='Points'))
```
